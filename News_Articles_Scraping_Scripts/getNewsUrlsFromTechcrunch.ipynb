```python
import requests
import pandas as pd
import datetime
import time

# üîê Your NewsAPI Key
API_KEY = "565ade5d80694d418b85a9bd8bcfc7c5"

# üîπ Storage for articles
techcrunch_articles = []

# üîπ Loop for the past 30 days (max 100 articles per day)
for day_offset in range(0, 30):
    date = (datetime.datetime.today() - datetime.timedelta(days=day_offset)).strftime('%Y-%m-%d')

    url = (
        f"https://newsapi.org/v2/everything?"
        f"sources=techcrunch&from={date}&to={date}&pageSize=100&language=en&apiKey={API_KEY}"
    )

    response = requests.get(url)
    data = response.json()

    if data["status"] == "ok":
        for article in data["articles"]:
            techcrunch_articles.append({
                "source_id": article["source"].get("id", "N/A"),
                "source_name": article["source"]["name"],
                "author": article.get("author", "N/A"),
                "title": article["title"],
                "description": article.get("description", "N/A"),
                "url": article["url"],
                "urlToImage": article.get("urlToImage", "N/A"),
                "publishedAt": article["publishedAt"],
                "content": article.get("content", "N/A"),
                "label": "real"
            })
    else:
        print(f"‚ùå Error fetching from TechCrunch on {date}: {data}")

    time.sleep(1)  # Avoid rate limit

# üîπ Save to CSV
df_tech = pd.DataFrame(techcrunch_articles)
df_tech.to_csv("techcrunch_news_3000.csv", index=False)

print(f"‚úÖ Saved {len(df_tech)} TechCrunch articles to techcrunch_news_3000.csv")
```
