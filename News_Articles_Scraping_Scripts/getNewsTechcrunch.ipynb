```python
import requests
from bs4 import BeautifulSoup

# TechCrunch article URL
url = "https://techcrunch.com/2025/03/31/streaming-service-plex-rolls-out-a-revamped-mobile-app/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Extract all paragraphs with the class wp-block-paragraph
paragraphs = soup.find_all("p", class_="wp-block-paragraph")
article_text = "\n\n".join([p.get_text(strip=True) for p in paragraphs])

if article_text:
    print("✅ Extracted article text:\n")
    print(article_text)
else:
    print("⚠️ Article content not found. Check selector.")
```
