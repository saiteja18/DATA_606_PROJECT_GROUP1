```python
import requests
from bs4 import BeautifulSoup

# CNN article URL
url = "https://www.cnn.com/2025/03/29/business/car-tariffs-prices-explained-dg/index.html"

# Send request with browser-like headers
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

# Extract paragraphs with exact class
paragraphs = soup.find_all("p", class_="paragraph inline-placeholder vossi-paragraph")

# Join all text content
cnn_article_text = "\n".join(p.get_text(strip=True) for p in paragraphs)

# Output the article
if cnn_article_text:
    print("✅ Full CNN Article Content:\n")
    print(cnn_article_text)
else:
    print("⚠️ Couldn't extract article. Check for dynamic loading or updated structure.")
```
