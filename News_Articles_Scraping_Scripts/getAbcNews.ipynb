```python
import requests
from bs4 import BeautifulSoup

# ABC News article URL
url = "https://abcnews.go.com/US/new-details-emerge-search-missing-university-pittsburgh-student/story?id=119641959"

# Set headers to mimic a real browser
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

# Parse HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Target div with article content using data-testid
article_div = soup.find("div", attrs={"data-testid": "prism-article-body"})

# Extract and print the full article
if article_div:
    paragraphs = article_div.find_all("p")
    full_article = "\n".join(p.get_text(strip=True) for p in paragraphs)
    print("✅ Full Article Content:\n")
    print(full_article)
else:
    print("⚠️ Article content not found. Check if 'data-testid' or structure has changed.")
```
