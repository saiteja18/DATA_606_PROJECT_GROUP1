```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

# Step 1: Load Excel file (uploaded in Colab)
df = pd.read_excel('/content/cnn_news_5000.xlsx')  # Replace with your actual filename

# Step 2: Initialize list to hold full article content
full_contents = []

# Step 3: Headers to mimic browser
headers = {'User-Agent': 'Mozilla/5.0'}

# Step 4: Loop through URLs and extract article content
for index, row in df.iterrows():
    url = row['url']  # Ensure this column exists in your Excel
    print(f"ğŸ” Scraping: {url}")
    try:
        response = requests.get(url, headers=headers, timeout=10)
        soup = BeautifulSoup(response.content, 'html.parser')

        # CNN class-based paragraph
        paragraphs = soup.find_all("p", class_="paragraph inline-placeholder vossi-paragraph")

        # Fallback if nothing is found
        if not paragraphs:
            paragraphs = soup.find_all("p")

        article_text = "\n".join(p.get_text(strip=True) for p in paragraphs)
        full_contents.append(article_text if article_text else "âš ï¸ No content found")

    except Exception as e:
        full_contents.append(f"âŒ Error: {str(e)}")

    # time.sleep(1)  # To prevent blocking

# Step 5: Add scraped content to the DataFrame
df["full_article_content"] = full_contents

# Step 6: Save to new Excel file
output_path = '/content/cnn_articles_with_full_content.xlsx'
df.to_excel(output_path, index=False)

print(f"âœ… Done! File saved to: {output_path}")
```
