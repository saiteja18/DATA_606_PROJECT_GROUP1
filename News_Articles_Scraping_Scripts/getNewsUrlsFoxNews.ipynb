```python
import requests
import pandas as pd
import datetime
import time

# üîê Your NewsAPI Key (replace with your key)
API_KEY = "18f06556ed6e4b90b1c6c90ce48b6b0d"

# üîπ Storage for articles
fox_articles = []

# üîπ Loop for the past 30 days
for day_offset in range(0, 30):
    date = (datetime.datetime.today() - datetime.timedelta(days=day_offset)).strftime('%Y-%m-%d')

    url = (
        f"https://newsapi.org/v2/everything?"
        f"sources=fox-news&from={date}&to={date}&pageSize=100&language=en&apiKey={API_KEY}"
    )

    response = requests.get(url)
    data = response.json()

    if data["status"] == "ok":
        for article in data["articles"]:
            fox_articles.append({
                "source_id": article["source"].get("id", "N/A"),
                "source_name": article["source"]["name"],
                "author": article.get("author", "N/A"),
                "title": article["title"],
                "description": article.get("description", "N/A"),
                "url": article["url"],
                "urlToImage": article.get("urlToImage", "N/A"),
                "publishedAt": article["publishedAt"],
                "content": article.get("content", "N/A"),
                "label": "real"  # Tagging as real news
            })
    else:
        print(f"‚ùå Error fetching from Fox News on {date}: {data}")

    time.sleep(1)  # Avoid API rate limit

# üîπ Save to CSV
df_fox = pd.DataFrame(fox_articles)
df_fox.to_csv("fox_news_3000.csv", index=False)

print(f"‚úÖ Saved {len(df_fox)} Fox News articles to fox_news_3000.csv")
```
