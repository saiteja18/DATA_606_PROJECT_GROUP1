```python
import requests
from bs4 import BeautifulSoup

url = "https://www.foxnews.com/us/north-carolina-woman-allegedly-murdered-green-beret-husband-concealed-his-death-officials"
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

article_body = soup.find("div", class_="article-body")

def is_valid_paragraph(p):
    text = p.get_text(strip=True)
    if not text:
        return False
    if p.find_parent(class_="caption") or p.find_parent(class_="image-ct"):
        return False
    if text.isupper():
        return False
    keywords_to_skip = [
        "SIGN UP", "CLICK HERE", "FOX NEWS APP", "NEWSLETTER",
        "SHARE THIS", "GET THE APP", "SUBSCRIBE", "RECOMMENDED"
    ]
    return not any(kw in text.upper() for kw in keywords_to_skip)

clean_paragraphs = []
if article_body:
    for p in article_body.find_all("p"):
        if is_valid_paragraph(p):
            clean_paragraphs.append(p.get_text(strip=True))

article_text = "\n\n".join(clean_paragraphs)
print(article_text)
```
